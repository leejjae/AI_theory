## RNN(Recurrent Neural Netwrok)

`-` RNN은 시퀀스 데이터(연속적인 데이터)를 처리하도록 만들어진 Neural
Network이다. 텍스트, 시계열자료(날씨, 시간에 따른 판매량..)등을 처리할
수 있다.

`-` Recurrent가 붙은 이유?  
$\rightarrow$ ’반복적인’이라는 뜻을 가진 Recurrent는 네트워크 구조에서
반복적으로 계산되기 때문이다.  
$t-1$시점에서의 cell에서 output이 $t$시점에서의 cell의 input으로
들어간다는 점에서, $t$ 시점에서의 값은 과거 값들에 영향을 받는다.
